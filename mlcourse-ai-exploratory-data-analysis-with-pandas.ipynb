```python
import pandas as pd
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python

data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python

x = (data['sex'] == 'Female')
round(data.loc[x,'age'].mean(),0)
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python

((data['native-country'] == 'Germany').sum() * 100  /  data.shape[0]).round(2)


#s.value_counts(normalize = True)
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python

r = data.loc[(data['salary'] == '>50K'),'age']
p = data.loc[(data['salary'] == '<=50K'),'age']
print(
    "The average age of the rich is {0} years,and the standard deviation is {1}".format(
        round(r.mean()),
        round(r.std(),2),
    )  
)
print('the average age of the poor is {0} years and the standard deviation is {1}'.format(
        round(p.mean()),
        round(p.std(),2),
    )  )
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
r = data['salary'] == '>50K'
data.loc[r,'education'].value_counts()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# You code here'
r = data.groupby(['race','sex'])
r['age'].describe()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# You code here
(data.loc[(data['sex'] == 'Male') & (data['marital-status'].str.startswith('Married')), 'salary'].value_counts(normalize=True) * 100).round(0)
```

```python
(data.loc[(data['sex']=='Male') & (data['marital-status'].isin(['Never-married', 
                                   'Separated', 
                                   'Divorced',
                                   'Widowed'])), 'salary'].value_counts(normalize=True) * 100).round(0)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_hour = data['hours-per-week'].max()
a = (data['salary'] == '>50K')
people_num = (data[(data['hours-per-week'] == max_hour)]).shape[0]
per = (data[(data['hours-per-week'] == max_hour) & (data['salary'] == '>50K')]).shape[0] / people_num * 100
print('the hours number is {0}, there are {1} people work such a number of hours and the percetnage of those who earn a lot among them are {2}% '.format( 
    max_hour,people_num,round(per)))
```
